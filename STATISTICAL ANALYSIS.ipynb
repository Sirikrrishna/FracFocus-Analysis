```python
def value_count(df, col):
    return df[col].value_counts()

def descriptive_continuous(df, col):
    mean = df[col].mean()
    med = df[col].median()
    mode = df[col].mode()
    sample_var = df[col].var()
    pop_var = df[col].var(ddof=0)
    maxi = df[col].max()
    mini = df[col].min()
    range = maxi - mini
    result = {"mean":mean, "median": med, "mode":mode, "Sample Variance":sample_var, "Population Variance":pop_var, "Maximum": maxi,"Minimum":mini, "Range":range}
    return result


def isnull_check(df, col):
    return df[col].isnull().sum()
```

```python
descriptive_continuous(df, 'TVD')
```

## Stastical Testing (ANOVA)

```python
df2=pd.read_csv("C://Users//sirik//Desktop//Chilary//FracFocusCSV//FracFocusRegistry_1.csv")
```

```python
df2.head()
```

```python
df2["TVD"].isnull().sum()
```

```python
df2["TVD"].isnull().sum()
```

```python
df2.shape
```

```python
from scipy.stats import f_oneway

group1 = df["TVD"]
group2 = df2["TVD"]
f_statistic, p_value = f_oneway(group1, group2)
print(p_value)
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. There is a significant difference between the means of the groups.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference between the means of the groups.")
```

### Based on the statistical testing conducted on the Total Vertical Depth (TVD) data spanning from 1955 to 2024, we can conclude that there is not a statistically significant change in the availability of oil and gas reserves at different surface depths over this time period.

### This suggests that there has been little to no variation in the distribution or accessibility of oil and gas resources in the subsurface layers during this timeframe.
